```python
''' Problem 1 '''

# Write a program that will accept input for the number of 25 centavos, 
# 1 peso, 5 pesos, 10 pesos, 20 pesos, 50 pesos, 100 pesos, 200 pesos
# 500 pesos, and 1000 pesos
# Compute and display the total amount of money

denominations = [1000, 500, 200, 100, 50, 20, 10, 5, 1, 0.25]
total = 0

print("Please enter the number of each denomination:")

# Calculating peso
for value in denominations[:-1]:
    numPeso = int(input(f"{value} peso{"s" if value != 1 else ""}: "))
    total += (value * numPeso)

# Calculating centavos
numCentavo = int(input("25 centavos: "))
total += (denominations[-1] * numCentavo)

# Displaying total amount
print("--------------------")
print(f"Total Amount: {total:.2f} pesos")
```

```python
''' Problem 2 '''

# Create a program that will accept Prelim, Midterm, Semi-Final, Final, Quiz, and Project Grade. 
# Compute the final grade: FG = (prelime x 15%) + (midterm x 15%) + (semifinal x 15%) + (final x 15%) + (quiz x 25%) + (project x 15%)
# Display a remark Congratulations! You passed the course if final grade is 75 and above; otherwise display Sorry, you failed

# Input grades to be calculated
print("Please enter the following grades:")
prelim = int(input("Prelims: "))
midterm = int(input("Midterms: "))
semifinal = int(input("Semi-finals: "))
final = int(input("Finals: "))
quiz = int(input("Quiz: "))
project = int(input("Project: "))

# Final grade formula
finalGrade = (prelim * 0.15) + (midterm * 0.15) + (semifinal * 0.15) + (final * 0.15) + (quiz * 0.25) + (project * 0.15)

# Display result and corresponding remark
print("--------------------")
print("Result:", finalGrade)
print("Remark: ", end = '')

if finalGrade >= 75:
    print("Congratulations! You passed the course.")
else:
    print("Sorry, you failed.")
```

```python
''' Problem 3 '''

# Accept an integer number.
# Compute and display the factorial of the number.

num = int(input("Enter a non-negative integer: "))

if num == 0: # zero factorial equals one
    factorial = 1
elif num > 0: # compute factorial of positive integers
    factorial = num
    for i in range(1, num):
        factorial *= num - i
else:
    print("Invalid input!")

# Display factorial result
print(f"{num}! = {factorial}")    
```

```python
''' Problem 4 '''

# Accept 10 Integer numbers.
# Count and display how many odd and even number input.

oddList = []
evenList = []

while True:
    integers = input("Enter ten integers each separated by a space: ").split(" ")

    if len(integers) != 10:
        print("Please enter exactly 10 integers.") # Handles insufficient integer inputs
        continue
    else:
        try:
            integers = [int(num) for num in integers] # Converts strings to integers
        except ValueError: 
            print("Please enter only integers.") # Handles non-integer inputs
            continue
        break

# Classifies each number into odd and even lists
for num in integers:
    if num % 2 == 0:
        evenList.append(num)
    else:
        oddList.append(num)

# Sort the odd and even list
oddList.sort()
evenList.sort()

# Displays the odd and even lists along with their counts
print(f"\nOdd Numbers: {', '.join(map(str, oddList))}")
print(f"Odd Count: {len(oddList)}")
print(f"Even Numbers: {', '.join(map(str, evenList))}")
print(f"Even Count: {len(evenList)}")
```

```python
''' Problem 5 '''

# Input a number and display if the number is a Prime Number or Not a Prime Number

while True:
    try:
        num = int(input("Enter a positive integer: "))

        if num < 1:
            print("Please enter only positive integers.")
            continue

        if num == 1:
            print(f"{num} is not prime.")
        elif num > 1:
            isPrime = True # Initializes flag to True
            for i in range(2, int(num ** 0.5) + 1):
                if num % i == 0:
                    isPrime = False # Sets flag to False
                    break

            if isPrime:
                print(f"{num} is prime.")
            else:
                print(f"{num} is not prime.")
        break
    except ValueError: # Handles non-positive integer inputs
        print("Invalid input. Please enter only integers.")
        continue
```

```python
''' Problem 6 '''

# Display a series of prime numbers based on user input.

while True:
    try:
        prime = [] # Initialize prime numbers list
        num = 2 # Initialized as 2 since 1 is already not prime
        num_of_prime = int(input("Enter number of prime numbers to dislay: "))

        if num_of_prime < 1: # Handles user inputs that are zero below
            print("Invalid input! Please enter only positive integers.")
            continue

        while len(prime) != num_of_prime: # Loop runs until length of prime list is equal to user input
            isPrime = True
            for i in range(2, int(num * 0.5) + 1): # Checks for factors between the number itself up to its square root
                if num % i == 0:
                    isPrime = False # Sets prime flag to False
                
            if isPrime:
                prime.append(num) # Add the number to the prime list
            num += 1 # Increment number to be checked

        print(f"Prime Numbers: {', '.join(map(str, prime))}") # Display prime list as string sequence
        break
    
    except ValueError: # Handles non-integer inputs
        print("Invalid input! Please enter only integers.")
        continue
```

```python
''' Problem 7 '''

# Create a program that will ask user to input the nth Fibonacci number series.
# Compute and display the Fibonacci number based on the input

while True:
    try:
        fibList = [0, 1] # Initialize Fibonacci list with zeroth and first terms
        num = int(input("Enter nth term in Fibonacci sequence to display: "))

        if num < 0: # Handles negative integer inputs
            print("Invalid input! Please enter only non-negative integers.")
            continue
        else:
            if num >= 2:
                for i in range(2, num + 1): # Loop starts at 2 since zeroth and first terms are already initialized
                    nFib = fibList[i - 1] + fibList[i - 2] # Adds previous two terms to get next term
                    fibList.append(nFib) # Adds new term to existing Fibonacci list

            nFib = fibList[num] # Gets nth term in the Fibonacci list according to user input

            print(f"F({num}) = {nFib}") # Displays nth term according to user input
            print(f"Fibonacci Sequence up to F({num}): {fibList}") # Displays Fibonacci sequence up to nth term
            break
    
    except ValueError: # Handles non-integer inputs
        print("Invalid input! Please enter only integers.")
        continue
```

```python
''' Problem 10 '''

# Create a Python program that will demonstrate list comprehension.

userInput = input("Enter integers each separated by a space: ").split()

numList = [int(num) for num in userInput]
divThree = [num for num in numList if num % 3 == 0]
squaredList = [(num ** 2) for num in numList]
cubedList = [(num ** 3) for num in numList]

print(f"Integer List: {numList}")
print(f"Divisible by 3: {divThree}")
print(f"Squared: {squaredList}")
print(f"Cubed: {cubedList}")
```

```python
''' Problem 11 '''

# Accept student name and grade and save it to a dictionary.
# Display the content of the dictionary.
# Display the name and grade of student with the highest and lowest grade.

grades = {} # Initialize empty dictionary for student names and grades
cont = 'y' # Initialize continue as 'yes'

while cont == 'y':
    studentName = input("Enter name of student: ").capitalize()
    
    while True:
        try:
            studentGrade = int(input(f"Enter grade of {studentName} (0-100 range): "))
            if studentGrade < 0 or studentGrade > 100:
                print("Invalid input! Please enter a grade between 0 and 100")
                continue
            else:
                break
        except ValueError: # Handles non-integer characters
            print("Invalid input! Please enter only integer values.")
            continue

    grades[studentName] = studentGrade # Adds student name and grade to dictionary

    while True:
        try:
            cont = input("Do you want to input for another student? (y/n): ").lower()
            if cont == 'y' or cont == 'n':
                break
            else:
                print("Invalid response! Please enter y/n.")
                continue
        except ValueError: # Handles non-alphabetic responses
            print("Invalid response! Please enter only y/n.")
            continue

print(f"\nStudent Grades: {grades}") # Displays the whole dictionary

# Gets highest and lowest grades in the dictionary
highGrade = max(grades.values())
lowGrade = min(grades.values())

# Initialize empty lists for student/s with highest and lowest grades
highStudent = []
lowStudent = []

for student in grades.keys(): # Handles multiple students that have the highest and lowest grades
    if grades[student] == highGrade:
        highStudent.append(student)
    elif grades[student] == lowGrade:
        lowStudent.append(student)

if len(grades) > 1: # Checks if there is more than one student in the dictionary
    print(f"Highest: {highStudent} - {highGrade}") # Display student/s with highest grade
    if len(set(grades.values())) != 1: # Checks if students do not have the same grade
        print(f"Lowest: {lowStudent} - {lowGrade}") # Display student/s with lowest grade
    else:
        print(f"Lowest: NA") # Display no lowest if students have the same grade
```

```python
''' Problem 13 '''

# Accept string input.
# Count vowel and consonant characters in the string.
# Display the string in reverse order.

vowels = 'aeiouAEIOU' # Initialize series lowercase and uppercase vowels as string
vowelCount = 0 
consonantCount = 0

string = input("Enter a line of string: ")
stringReversed = ''.join(reversed(string)) # Creates a reversed version of the string input

for char in string: # Iterates through each character of the string input
    if char.isalpha(): # Checks if character is an alphabetic character
        if char in vowels: # Checks if character is a vowel
            vowelCount += 1
        else:
            consonantCount += 1

# Displays results
print(f"Vowel Occurences: {vowelCount}")
print(f"Consonant Occurences: {consonantCount}")
print(f"Reversed: {stringReversed}")
```

```python
''' Problem 14 '''

# Accept string input.
# Display if the string is Palindrome or Not a Palindrome

string = input("Enter a line of string: ")
stringReversed = ''.join(reversed(string)) # Reversed version of string input

if string.lower() == stringReversed.lower(): # Uses lowercase for case insensitivity
    print(f"{string} is a palindrome!")
else:
    print(f"{string} is not a palindrome.")
```

```python
''' Problem 15 '''

# Create a Python program that will use at least 3 user-defined function. Demonstrate function with return value and void function.

def greet(name): # This function displays a greeting and returns None
    print(f"Hi {name}! This is a simple calculator program.")

def calculate(x, y, op): # This function makes arithmetic calculations and returns the solution result
    if op == '+':
        return x + y
    elif op == '-':
        return x - y
    elif op == '*':
        return x * y
    elif op == '/':
        return x / y
    else:
        return "Invalid operation! Please select from [+-*/]."
    
def main(): # This functions accepts the main user input and displays the calculation output
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    operation = input("Enter operation [+-*/]: ")
    result = calculate(num1, num2, operation)
    print(f"Result: {result}")

proceed = 'y'
name = input("Enter your name: ")

greet(name)

while proceed == 'y':
    main()
    proceed = input("Do you want to make another calculation? (y/n): ").lower()
```

```python
''' Problem 16 '''

# Create a Python program that will demonstrate lambda

add = lambda x,y: x + y
subtract = lambda x,y: x - y
multiply = lambda x,y: x * y
divide = lambda x,y: x / y if y != 0 else "Error! Divided by zero."

proceed = 'y'
while proceed == 'y':
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    operation = input("Enter operation ( + - * / ): ")

    if operation == '+':
        result = add(num1, num2)
    elif operation == '-':
        result = subtract(num1, num2)   
    elif operation == '*':
        result = multiply(num1, num2)    
    elif operation == '/':
        result = divide(num1, num2)
        if result != "Error":
            result = round(result, 2)
            
    print (f"Result: {result}")

    proceed = input("Do you want to make another calculation? (y/n): ").lower()
```
